## REGRESIÓN LINEAL

La regresión lineal es un tipo de modelo de aprendizaje que se encarga de predecir una variable, llamada variable objetivo y comúnmente denominada con la letra "y", en base a otras variables, llamadas variables predictoras, referidas como "x". En este tipo de modelos, se busca trazar una línea recta, que mejor se ajuste a los puntos de "y", en relación a "x". 



La regresión suele ser usada como benchmark en modelos de análisis y predicción de datos, ya que es muy simple de realizar, y normalmente tiene buenos resultados si la información que se le provee al modelo es buena. Sin emabrgo, pueden surgir varios problemas cuando los datos son más complejos, y tienen tendencias distintas a una lineal. También puede pasar, que cuando existen más de una variable predictora, la manera en que se complementan entre ellas genere problemas para que el modelo sea preciso o confiable.

### PROBLEMAS POTENCIALES EN REGRESIÓN LINEAL 

- Falta de Linealidad en la respuesta

Si la tendencia entre la variable predictora y predicha sigue una tendencia distinta a una lineal, es muy posible que el modelo no pueda analizar correctamente la tendencia, y por ende, no la capte y el modelo no funcione para predecir o analizar. 


En casos así, se recomienda utilizar un modelo distinto, capaz de captar estas tendencias distintas, como una regresión polinomial, que puede adecuarse mejor a este tipo de escenarios. 

- Correlación en Residuos

La regresión lineal, por Mínimos Cuadrados, asume que los residuos son Independientes e Idénticamente distribuidos, esto significa que de haber autocorrelación en los residuos, se viola esta asunción. Es decir, si existe correlación en los residuos, entonces hay también tendencias o patrones que deberían modelarse con el modelo. [1]


- Varianza no constante en Residuos

También llamada heterocedasticidad, indica que la dispersión de los errores no es uniforme, a lo largo de los valores de las variables predictoras. Esto viola los supuestos sobre los cuales se construye la regresión lineal. Esto, puede producir que los p-values y los valores de confianza se calculen de manera inexacta, y por ende el modelo no sea tan confiable como pudiera parecer. [2]

-  Outliers



Los outliers son datos dentro de la información que recibe el modelo, que se encuentran fuertemente desviados del patrón de los mismos. Es decir, si los valores posibles de "y" para un dado valor "x", se puede denotar en un cierto rango, un outlier se posicionará fuera de este rango de manera exagerada o muy notable. Se pueden deber a diversos factores, y normalmente, en regresión, sesgan los coeficientes del modelos para, inclinando la recta hacia ellos a pesar de que quizás no suponen información muy importante para el modelo. [3] [5]

- High-Leverage Points



Los high-leverage points, son similares a los outliers, pero en lugar de tener valores atípicos para "y", tienen valores atípicos para "x", lo cual supone un problema en el ajuste del modelo. Es importante analizar si estos valores se deben a una mala medición en alguna muestra, y si no, se puede hacer más robusto el modelo para que penalice errores significativamente más grandes.[3] [5]

- Colinealidad en los datos

Existe cuando dos o más variables predictoras no son independientes entre sí, es decir, cuando existe información o tendencias iguales o muy similares en más de una variable. Esto genera inestabilidad en los coeficientes del modelo, ya que capta dos veces la misma información, y es difícil aislar el comportamiento de cada variable por separado. Esto se puede amortiguar con técnicas como regularización o con un análisis del VIF (Variance Inflation Factor) [4]

## ÁRBOL PARA REGRESIÓN

Un árbol para regresión es un tipo de modelo que predice valores numéricos de manera similar a un arbol de clasificación. Divide los datos a partir de características importantes, empezando por una pregunta raíz para posteriormente desarrollarse en distintas ramas. De esta manera, va subdividiendo los datos para llegar a una conclusión en los últimos nodos del modelo, llamados hojas. Estos modelos son útiles cuando tus datos se benefician de irse categorizando o subdividiendo por características. [6]

## ÁRBOL PARA CLASIFICACIÓN

Un árbol de clasificación, es un modelo que se encarga, a traves de un proceso de subdivisiones, de categorizar entradas de datos para predecir una clase. Para ello, utiliza una estructura jerárquica que consta de un nodo raíz, ramas, nodos internos y nodos de hoja. El funcionamiento es muy similar al de un árbol de regresión, solo que su resultado final no es un valor numérico, si no una clase. 



En este tipo de modelos, se busca que la varianza entre cada subdivisión generada sea la misma, y existen distintas métricas que se pueden minimizar a la hora de elegir las mejores subdivisiones para cada rama y nodo. A su vez, existe un proceso de "podado" de árbol, que consiste en simplificar lo más posible la cantidad de decisiones que toma el modelo, manteniéndolo lo más simple y eficiente posible.

[7]

## BOOTSTRAP

Bootstrap se refiere a un método de re-muestreo de un conjunto de datos, que tiene el objetivo de mejorar la confianza estadística de los parámetros que obtenemos de dicho conjunto de datos, simulando el proceso de muestreo original que se lleva a cabo cuando se obtienen. El bootstrap consiste en re-muestrear la muestra original, con reemplazo, ya que normalmente es más eficiente que tomar muestras nuevas para cada estudio. Al hacer esto, podemos hacer más robusta nuestra estimación de intervalos de confianza para los parámetros de un modelo, y por ende validarlo estadísticamente de una manera más eficiente. [8]

## ENSAMBLES DE MODELOS

El ensamble de modelos se refiere al método en aprendizaje estadístico que se encarga de combinar uno o más modelos, con el objetivo de hacer un mejor modelo final, el cual se beneficia del aprendizaje de cada uno de ellos, y mantiene puntos positivos de cada tipo de modelo que otro probablemente no tenga. Es decir, combina modelos para utilizar los puntos fuertes de cada uno y hacer una combinación que los aproveche todos. 



Este ensamble de modelos está presente, por ejemplo, en las redes neuronales, ya que distintos modelos se alimentan entre sí de manera secuencial para producir un resultado más robusto. Un ensamble paralelo entrena los modelos de manera aislada, mientras que los secuenciales los entrena también entre ellos, lo cual da lugar a distintos resultados. 

Esta técnica ayuda a minimizar problemas de modelado como el bias-variance tradeoff, famoso por que no puedes tener uno sin el otro, y debes decidir cómo balancear cada uno.

[9]

## BAGGING

Una técnica famosa de ensamble de modelos es el bagging (boostrsap aggregating). Consiste en aprovechar la naturaleza del re-muestreo aleatorio del bootstrap, para generar "n" número de sub-muestras distintas, todas provenientes del mismo conjunto. Con esto, tenemos una cantidad de muestras importantes, con las cuales posteriormente podemos entrenar modelos con el mismo algoritmo, pero distintas muestras de entrenamiento, para luego ensamblar los modelos y obtener resultados más representativos, lo cual es muy útil cuando por ejemplo, los conjuntos iniciales de datos son reducidos, o es muy costoso obtener más datos. 



[9] [10]

## BOOSTING

Boosting es otra técnica de ensamble de modelos, parecida al bagging. Sin embargo, el boosting en vez de ser entrenado de forma paralela como el bagging, se entrena de manera secuencial, ya que se entrena cada modelo con los errores del anterior, con el objetivo de no equivocarse donde los anteriores sí. 



De esta manera, cada modelo va aprendiendo de los anteriores, y por ende se va mejorando con cada iteración. Sin embargo, puede volverse sensible ya que se adapta fuertemente a los datos, y cosas como outliers pueden perjudicar el rendimiento del mismo. 

[11]

# ANÁLISIS DEL DATASET

- ¿De dónde viene?

Los datos proviene del sitio web Kaggle, donde se publican diferentes concursos y conjuntos de datos para hacer análisis y generar modelo desde distintos ámbitos. 

- ¿Qué contiene?

Contiene medidas física y morfológicas de plantas recolectadas en un estudio experimental que compara invernadores con tecnología IoT y aquellos sin ella. Contiene 30,000 muestras con 14 columnas, representando distintas métricas de crecimiento, niveles de clorofila y otros datos característicos del crecimento de las plantas.

- ¿Qué información dan las muestras?

1. Random: 
2. ACHP: (Average Clorophyll Content per Plant) La cantidad media de clorofila por planta, que mide la actividad fotosintética de la misma. 
3. PHR: (Plant Height Rate) Tasa de Crecimiento de Altura correspondiente a cada planta, mide cuánto crece a lo largo del tiempo. 
4. AWWGV: (Average Wet Weight of Vegetative Growth) El total del peso de las partes húmedas y frescas de las partes de la planta por encima del suelo. 
5. ALAP: (Average Leaf Area per Plant) Promedio del área de las superficies de las hojas de las plantas que impactan la fotosíntesis.
6. ANPL: (Average Number of Leaves per Plant) Cantidad promedio de hojas por planta, que indican la madurez de las plantas y la densidad de su follaje.
7. ARD: (Average Root Diameter) Diámetro medio de la raíz, grosor de las raíces que se relacionan con la nutrición que reciben.
8. ADWR: (Average Dry Weight of Roots) peso medio de las raíces de la planta luego de secarse.
9. PDMVG: (Percentage of Dry Matter in Vegetative Growth) Porcentaje de materia seca en crecimiento vegetativo, mide el contenido sólido en las muestras. 
10. ARL: (Average Root Lenght) Longitud media de las raíces de la planta, indica el desarrollo de las raíces y su profundidad. 
11. AWWR: (Average wet weight of roots) Peso promedio de las raíces humidificadas.
12. ADWV: (Average dry weight of vegetative parts) Peso promedio de las partes vegetativas secas de la planta.
13. PDMRG: (% dry matter in root growth) Porcentaje de materia seca en el crecimiento de las raíces. 
14. Class: Etiqueta de cada grupo experimental (SA, SB, SC, TA, TB, TC).
 

- ¿Qué se quiere analizar?

Se busca analizar el impacto de cada una de estas características en el contenido promedio de clorofila por planta, ya que funciona como un buen indicador de la actividad fotosintética de la planta, y por ende nos sirve para medir cómo se relaciona cada factor con la capacidad fotosintética de las plantas. 


- ¿Qué resultado se podría encontrar al ajustar un modelo?

Se podría ver de manera clara, con qué características se puede estimar de manera más eficaz la variable objetivo, y por ende, durante el crecimiento de las plantas, buscar maximizar aquella característica que se muestra positiva a los resultados que buscamos producir.

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PowerTransformer
from sklearn.preprocessing import OneHotEncoder

from xgboost import XGBRegressor

from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import ElasticNet
from sklearn.ensemble import RandomForestRegressor

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import r2_score

from sklearn.decomposition import PCA

import optuna
```

#### PRESCRICPCIÓN DE LOS DATOS

```python
df = pd.read_csv("Greenhouse Plant Growth Metrics.csv")
df
```

```python
def bivariate_analysis(df, target):
    
    for col in df.columns:
        if col == target:
            continue
        
        print(f"\n=== Bivariate Analysis: {col} vs {target} ===")
        
        if pd.api.types.is_numeric_dtype(df[col]) and pd.api.types.is_numeric_dtype(df[target]):
            # Numeric vs Numeric
            corr = df[[col, target]].corr().iloc[0,1]
            print(f"Correlation (Pearson): {corr:.4f}")
            sns.scatterplot(x=col, y=target, data=df)
            plt.title(f'Scatter Plot: {col} vs {target}')
            plt.show()
        
        elif pd.api.types.is_numeric_dtype(df[col]) and not pd.api.types.is_numeric_dtype(df[target]):
            # Numeric feature vs Categorical target
            sns.boxplot(x=target, y=col, data=df)
            plt.title(f'Box Plot: {col} by {target}')
            plt.show()
        
        elif not pd.api.types.is_numeric_dtype(df[col]) and pd.api.types.is_numeric_dtype(df[target]):
            # Categorical feature vs Numeric target
            group_mean = df.groupby(col)[target].mean().sort_values(ascending=False)
            print("Mean target by category:")
            print(group_mean)
            sns.barplot(x=group_mean.index, y=group_mean.values)
            plt.title(f'Mean {target} by {col}')
            plt.xticks(rotation=45)
            plt.ylabel(f'Mean {target}')
            plt.xlabel(col)
            plt.show()
        
        else:
            # Categorical vs Categorical
            ct = pd.crosstab(df[col], df[target], normalize='index')
            print("Cross-tabulation (normalized):")
            print(ct)
            ct.plot(kind='bar', stacked=True)
            plt.title(f'Stacked Bar Chart: {col} vs {target}')
            plt.xticks(rotation=45)
            plt.show()
```

```python
bivariate_analysis(df, "ACHP")
```

```python
numerical_features = ["PHR", "AWWGV", "ALAP", "ANPL", "ARD", "ADWR", "PDMVG", "ARL", "AWWR", "ADWV", "PDMRG"]
categorical_features = ["Class", "Random"]
```

```python
x = df.copy()
x = x.drop(columns="ACHP")

y = df["ACHP"]
```

```python
numerical_transformer = PowerTransformer(method='box-cox', standardize=True)

categorical_transformer = OneHotEncoder()

pca = PCA(n_components=0.95)

preprocessor = ColumnTransformer(
    transformers = [
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)
```

## REGRESIÓN LINEAL

```python
model = ElasticNet()

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('pca', pca),  
    ('model', model)
])
```

```python
from sklearn.model_selection import cross_val_score

cv = 10
scores = cross_val_score(pipeline, x, y, scoring = "r2", cv=cv)
r2s = scores

r2_mean = r2s.mean()
r2_std = r2s.std()

print(r2_mean, r2_std)
```

### OPTIMIZACIÓN DE HIPERPARÁMETROS

```python
# Definir función a optimizar
def objective(trial):
    # Sugerir parámetros
    alpha = trial.suggest_loguniform("alpha", 1e-4, 100.0)
    l1_ratio = trial.suggest_uniform("l1_ratio", 0.0, 1.0)

    # Definir el modelo
    model = ElasticNet(alpha=alpha, 
                       l1_ratio=l1_ratio, 
                       max_iter=10000)
    
    pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('pca', pca),  
    ('model', model)
])

    # Realizar cv y obtener el score promedio
    score = cross_val_score(pipeline, x, y, cv=10, scoring="r2", n_jobs=-1).mean()
    
    return score  

#OPTUNA
study = optuna.create_study(direction="maximize")  # Maximize r2
study.optimize(objective, n_trials=50)  # Try n different sets of hyperparameters


print("Best hyperparameters:", study.best_params)
```

```python
alpha = 0.19025625301003826
l1_ratio = 0.05013143795505734
```

### MODELO PROPUESTO

```python
regression = ElasticNet(alpha = 0.19025625301003826, 
                       l1_ratio = 0.05013143795505734, 
                       max_iter=10000)

pipeline_regression = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('pca', pca),  
    ('model', regression)
])
```

### DIAGNÓSTICO DE RESIDUALES

```python
pipeline_regression.fit(x, y)
y_pred_regression = pipeline_regression.predict(x)
errores_regression = y - y_pred_regression

plt.scatter(y_pred_regression, errores_regression, c="darkblue")
```

```python
linreg_regression = LinearRegression(fit_intercept = True)

errores_regression = np.reshape(errores_regression, [-1,1])
y_pred_regression = np.reshape(y_pred_regression, [-1,1])

linreg_regression.fit(y_pred_regression, errores_regression)

linreg_pred_regression = linreg_regression.predict(y_pred_regression)

plt.scatter(y_pred_regression, errores_regression, c="darkblue")
plt.plot(y_pred_regression, linreg_pred_regression, color='black', linewidth=2, label='Línea de regresión') 
```

## ÁRBOL DE REGRESIÓN

```python
model = DecisionTreeRegressor()

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('pca', pca),  
    ('model', model)
])
```

```python
from sklearn.model_selection import cross_val_score

cv = 10
scores = cross_val_score(pipeline, x, y, scoring = "r2", cv=cv)
r2s = scores

r2_mean = r2s.mean()
r2_std = r2s.std()

print(r2_mean, r2_std)
```

### OPTIMIZACIÓN DE HIPERPARÁMETROS

```python
# Definir función a optimizar
def objective(trial):
    # Sugerir parámetros
    max_depth = trial.suggest_int("max_depth", 2, 30)
    min_samples_split = trial.suggest_int("min_samples_split", 2, 20)
    min_samples_leaf = trial.suggest_int("min_samples_leaf", 1, 20)
    max_features = trial.suggest_categorical("max_features", ["sqrt", "log2", None])
    ccp_alpha = trial.suggest_float("ccp_alpha", 0.0, 0.1)

    # Definir el modelo
    model = DecisionTreeRegressor(
                        min_samples_split=min_samples_split,
                        min_samples_leaf=min_samples_leaf,
                        max_depth=max_depth,
                        max_features=max_features,
                        ccp_alpha=ccp_alpha)
    
    pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('pca', pca),  
    ('model', model)
])

    # Realizar cv y obtener el score promedio
    score = cross_val_score(pipeline, x, y, cv=10, scoring="r2", n_jobs=-1).mean()
    
    return score  

#OPTUNA
study = optuna.create_study(direction="maximize")  # Maximize r2
study.optimize(objective, n_trials=50)  # Try n different sets of hyperparameters


print("Best hyperparameters:", study.best_params)
```

### OPTIMIZACIÓN DE HIPERPARÁMETROS

```python
# Definir función a optimizar
def objective(trial):
    # Sugerir parámetros
    n_estimators = trial.suggest_int("n_estimators", 1, 1000)  
    max_depth = trial.suggest_int("max_depth", 1, 1000)  
    max_leaves = trial.suggest_int("max_leaves", 1, 1000) 
    learning_rate = trial.suggest_loguniform("learning_rate", 1e-4, 0.3) 

    # Definir el modelo
    model = XGBRegressor(n_jobs=-1, 
                         n_estimators=n_estimators, 
                         max_leaves=max_leaves, 
                         max_depth=max_depth,
                         learning_rate=learning_rate,  
                         random_state=42)
    
    pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('pca', pca),  
    ('model', model)
])

    # Realizar cv y obtener el score promedio
    score = cross_val_score(pipeline, x, y, cv=10, scoring="r2", n_jobs=-1).mean()
    
    return score  

#OPTUNA
study = optuna.create_study(direction="maximize")  # Maximize r2
study.optimize(objective, n_trials=50)  # Try n different sets of hyperparameters


print("Best hyperparameters:", study.best_params)
```
